```python
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np
```

# Linear Graph

```python
plt.plot([1,2,3],[5,7,8])
plt.title("first graph")
plt.ylabel('Y axcis')
plt.xlabel('x axcis')
```

```python
style.use('ggplot')
x = [1,2,3,4]
y = [8,7,2,6]
x1 = [3, 6, 8, 9]
y1 = [2, 5, 7, 3]
plt.plot(x,y,'g', label='line', linewidth=2)
plt.plot(x1,y1,'r', label='line1', linewidth=2)
plt.title('X and Y')
plt.ylabel("value of y")
plt.xlabel("value of x")
plt.legend()
plt.grid(True, color='k')
plt.show()
```

# Bar Graph

```python
x = [1, 2, 3, 4]
y = [22, 7, 9, 11]
plt.bar(x,y,label='bar1', color='g', width=0.3)
plt.title('Bar Graph')
plt.ylabel("value of y")
plt.xlabel("value of x")
plt.legend()
plt.show()
```

```python
x = [1, 2, 4, 8]
y = [3, 7, 9, 11]
x1 = [3, 6, 8, 9]
y1 = [2, 5, 7, 3]
plt.bar(x1,y1, label='bar2', color='r')
plt.bar(x,y,label='bar1', color='g')
plt.title('Bar Graph')
plt.ylabel("value of x")
plt.xlabel("value of y")
plt.legend()
plt.show()
```

# Histogram

```python
x = [95, 64, 55, 80, 60, 57, 65, 67, 63, 72, 68, 70, 74, 65, 95, 64, 55, 80, 60, 57, 65, 67, 63, 72, 68, 70, 74, 65]
y = [55, 62, 69, 76, 83, 90, 97]
plt.hist(x,y, histtype='bar',rwidth=0.99)
plt.title('Histogram')
plt.ylabel("value of x")
plt.xlabel("value of y")
plt.show()
```


# Scatter Diagram

```python
x = [1, 2, 3, 4]
y = [22, 7, 9, 11]
plt.scatter(x,y, color='k')
plt.title('scatter diagram')
plt.ylabel("value of y")
plt.xlabel("value of x")
plt.show()
```

# Pi chart

```python
x = [9, 5, 5, 3]
l = ['1st', '2nd', '3rd', '4th']
plt.pie(x, labels=l, colors=('c','m','r','b'), startangle=90, shadow=True, explode=(0,0,0.1,0), autopct='%1.2f%%')
plt.title('pi chart')
plt.show()
```

# Showing two plot in one graph

```python
x = np.arange(0,5, 0.1)
y = np.arange(0,5,0.1)
x1 = np.arange(0,5,0.2)
y1 = np.arange(0,5, 0.2)
plt.subplot(211)
plt.plot(x,y)
plt.title('plot1')
plt.subplot(212)
plt.plot(x1,y1)
plt.title('plot2')
plt.ylabel('value of x')
plt.xlabel('value of y')
plt.show()
```

# Frequency polygon

```python
y = [6, 12, 6, 2, 0, 2]
x = [55, 62, 69, 76, 83, 90]
plt.plot(x,y)
plt.title('frequency polygon')
plt.ylabel("value of y")
plt.xlabel("value of x")
plt.show()
```
